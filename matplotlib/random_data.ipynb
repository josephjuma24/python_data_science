```python
from matplotlib import pyplot as plt
import numpy as np

# Generate 100 random data points along 3 dimensions
# np.random.randn creates normally distributed random numbers with mean 0 and variance 1

x, y, scale = np.random.randn(3, 100)
fig, ax = plt.subplots()

# Map each onto a scatterplot we'll create with Matplotlib
# x and y are the coordinates, c is the color (based on scale value)
# s is the size of each point (absolute value of scale multiplied by 500)

ax.scatter(x=x, y=y, c=scale, s=np.abs(scale)*500)
ax.set(title="Scatter Plot of Some random data!")
       
plt.show() # Display the figure
```

  <li><code>x, y, scale = np.random.randn(3, 100)</code>
