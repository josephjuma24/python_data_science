```python
import matplotlib.pyplot as plt
import numpy as np

def plot_scatter(ax, prng, nb_samples=100):
    for mu, sigma, marker in [(0, 1, 'o'), (1, 2, 's')]:
        x = mu + sigma * prng.randn(nb_samples)
        y = mu + sigma * prng.randn(nb_samples)
        ax.plot(x, y, ls='none', marker=marker)

    ax.set_xlabel("X-label")
    return ax
 
```

    <ul>
      <li>Creates a new Matplotlib figure.</li>

    </ul>

  <li><code>plot_scatter(ax, prng)</code>

<h3>Summary</h3>

```python
prng = np.random.RandomState(42)
fig, ax = plt.subplots()
plot_scatter(ax, prng)
plt.show()
```

<h3>Scatter Plot Interpretation</h3>
